```python
#Extracting financial data of Fidelity Bank from Edgar
import requests
import pandas
headers = {'user-agent': 'sample text'}
url = "https://www.sec.gov/Archives/edgar/data/0001031233/000156459021016418/ppbn-10k_20201231.htm" 
financial_report = requests.get(url,headers=headers)
df =pandas.read_html(financial_report.text)
df = df[12].dropna(how = 'all')

df = df.fillna(0)
df
```
